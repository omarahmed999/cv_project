```python
import cv2
import numpy as np

def preprocess_image(image, target_size=(224, 224)):
    """
    Resize and normalize the image.
    - image: Input image as a NumPy array (BGR)
    - target_size: Tuple (width, height)
    Returns: Preprocessed image (float32 normalized)
    """
    # Resize
    resized = cv2.resize(image, target_size)
    
    # Normalize to [0, 1]
    normalized = resized.astype(np.float32) / 255.0
    
    return normalized
```

```python
import cv2
import numpy as np

def adjust_contrast(image, alpha=1.5, beta=0):
    """
    Adjust contrast and brightness of an image.
    alpha: Contrast control (1.0 means no change, higher means more contrast)
    beta: Brightness control (0 means no change)
    """
    return cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Example usage
image = cv2.imread('images/maksssksksss1.png')
image_contrast = adjust_contrast(image)
cv2.imshow('Adjusted Contrast', image_contrast)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def threshold_image(image, thresh_value=127):
    """
    Apply binary thresholding to an image.
    The image is converted into a binary image where pixel values above
    the threshold are set to 255, and those below are set to 0.
    """
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    _, thresh_img = cv2.threshold(gray, thresh_value, 255, cv2.THRESH_BINARY)
    return thresh_img

# Example usage
image = cv2.imread('images/maksssksksss1.png')
thresh_img = threshold_image(image)
cv2.imshow('Thresholded Image', thresh_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def apply_blur(image, kernel_size=(5, 5)):
    """
    Apply Gaussian blur to an image.
    kernel_size: Size of the kernel (must be odd values like (3, 3), (5, 5))
    """
    return cv2.GaussianBlur(image, kernel_size, 0)

# Example usage
image = cv2.imread('images/maksssksksss1.png')
blurred_image = apply_blur(image)
cv2.imshow('Blurred Image', blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def apply_sharpening(image):
    """
    Apply sharpening filter to the image.
    Uses a kernel to highlight edges.
    """
    kernel = np.array([[0, -1, 0],
                       [-1, 5,-1],
                       [0, -1, 0]])  # Sharpening kernel
    return cv2.filter2D(image, -1, kernel)

# Example usage
image = cv2.imread('images/maksssksksss1.png')
sharpened_image = apply_sharpening(image)
cv2.imshow('Sharpened Image', sharpened_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def preprocess__image(image):
    # Resize
    image = cv2.resize(image, (224, 224))
    
    # Contrast Adjustment
    image = adjust_contrast(image, alpha=1.5, beta=0)
    
    # Thresholding
    image = threshold_image(image, thresh_value=127)
    
    # Blurring
    image = apply_blur(image, kernel_size=(5, 5))
    
    # Sharpening
    image = apply_sharpening(image)
    
    # Normalize
    image = image.astype(np.float32) / 255.0
    
    return image

# Example usage
image = cv2.imread('images/maksssksksss2.png')
processed_image = preprocess_image(image)
cv2.imshow('Processed Image', processed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def segment_image(image, method="thresholding"):
    """
    Apply different segmentation methods to an image.
    Args:
        image: The input image to segment.
        method: The segmentation method to use. Options: "thresholding", "edges", "region", "clustering".
    Returns:
        segmented_image: The segmented image after applying the selected method.
    """
    # Convert the image to grayscale for easier processing
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    if method == "thresholding":
        # Apply Otsu's thresholding
        _, segmented_image = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    elif method == "edges":
        # Apply Canny edge detection
        segmented_image = cv2.Canny(gray, 100, 200)

    elif method == "region":
        # Apply morphological operations for region-based segmentation
        kernel = np.ones((5,5), np.uint8)
        # First, apply a threshold to highlight regions of interest
        _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
        # Then, use morphological closing (dilation followed by erosion) to fill gaps
        segmented_image = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

    elif method == "clustering":
        # Use K-means clustering for color-based segmentation
        Z = image.reshape((-1, 3))
        Z = np.float32(Z)

        # Define criteria and apply KMeans clustering
        criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
        K = 3  # Number of clusters (you can adjust this based on your need)
        ret, label, center = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

        # Convert back to 8 bit values
        center = np.uint8(center)
        segmented_image = center[label.flatten()]
        segmented_image = segmented_image.reshape(image.shape)

    else:
        raise ValueError("Invalid method. Choose from 'thresholding', 'edges', 'region', 'clustering'.")
    
    return segmented_image

def visualize_segmentation(image, method):
    """
    Visualizes the original image and the segmented result.
    Args:
        image: The original image.
        method: The segmentation method applied.
    """
    segmented = segment_image(image, method)
    
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(segmented, cmap='gray')
    plt.title(f"Segmented Image ({method})")
    plt.axis('off')
    
    plt.show()
```

```python
import os
import cv2
import xml.etree.ElementTree as ET
import numpy as np

# Class names (map them to indices later)
CLASSES = ['with_mask', 'without_mask', 'mask_weared_incorrect']

def preprocess_image(image, target_size=(224, 224)):
    resized = cv2.resize(image, target_size)
    normalized = resized.astype(np.float32) / 255.0
    return normalized

def load_dataset(images_path='images', annotations_path='annotations', target_size=(224, 224)):
    X = []  # Preprocessed images
    y = []  # Corresponding bounding boxes and labels

    image_files = [f for f in os.listdir(images_path) if f.endswith('.png')]

    for img_file in image_files:
        xml_file = img_file.replace('.png', '.xml')
        xml_path = os.path.join(annotations_path, xml_file)

        if not os.path.exists(xml_path):
            continue  # Skip if annotation missing

        # Load and preprocess image
        image_path = os.path.join(images_path, img_file)
        image = cv2.imread(image_path)
        preprocessed = preprocess_image(image, target_size)
        X.append(preprocessed)

        # Parse XML annotation
        boxes = []
        tree = ET.parse(xml_path)
        root = tree.getroot()

        for obj in root.findall('object'):
            label = obj.find('name').text
            bbox = obj.find('bndbox')
            x_min = int(bbox.find('xmin').text)
            y_min = int(bbox.find('ymin').text)
            x_max = int(bbox.find('xmax').text)
            y_max = int(bbox.find('ymax').text)
            boxes.append([label, x_min, y_min, x_max, y_max])
        
        y.append(boxes)

    return X, y
```

```python
def load_dataset(images_path='images', annotations_path='annotations', target_size=(224, 224)):
    X = []
    y = []

    image_files = [f for f in os.listdir(images_path) if f.endswith('.png')]

    for img_file in image_files:
        xml_file = img_file.replace('.png', '.xml')
        xml_path = os.path.join(annotations_path, xml_file)

        if not os.path.exists(xml_path):
            continue

        # Load original image
        image_path = os.path.join(images_path, img_file)
        image = cv2.imread(image_path)
        orig_h, orig_w = image.shape[:2]

        # Resize and normalize
        preprocessed = preprocess_image(image, target_size)
        X.append(preprocessed)

        # Parse and normalize bounding boxes
        boxes = []
        tree = ET.parse(xml_path)
        root = tree.getroot()

        for obj in root.findall('object'):
            label = obj.find('name').text
            class_id = CLASSES.index(label)  # Convert label to integer

            bbox = obj.find('bndbox')
            x_min = int(bbox.find('xmin').text)
            y_min = int(bbox.find('ymin').text)
            x_max = int(bbox.find('xmax').text)
            y_max = int(bbox.find('ymax').text)

            # Normalize box to [0, 1] relative to original size
            x_min_norm = x_min / orig_w
            y_min_norm = y_min / orig_h
            x_max_norm = x_max / orig_w
            y_max_norm = y_max / orig_h

            boxes.append([class_id, x_min_norm, y_min_norm, x_max_norm, y_max_norm])

        y.append(boxes)

    return np.array(X), y
```

```python

def build_object_detection_model(input_shape=(224, 224, 3), num_classes=3):
    inputs = layers.Input(shape=input_shape)

    # Feature extractor
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(inputs)
    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(64, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(128, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Flatten()(x)
    x = layers.Dense(128, activation='relu')(x)

    # Output layer
    class_output = layers.Dense(num_classes, activation='softmax', name='class_output')(x)
    bbox_output = layers.Dense(4, activation='sigmoid', name='bbox_output')(x)  # [x_min, y_min, x_max, y_max]

    model = models.Model(inputs=inputs, outputs=[class_output, bbox_output])
    return model
```

```python
model = build_object_detection_model()
model.compile(
    optimizer='adam',
    loss={
        'class_output': 'sparse_categorical_crossentropy',
        'bbox_output': 'mse'  # bounding box regression
    },
    metrics={
        'class_output': 'accuracy',
        'bbox_output': 'mse'
    }
)
model.summary()
```

```python
def prepare_labels(y):
    """
    Convert y from list of lists to:
    - class_labels: shape (num_samples,)
    - bbox_labels: shape (num_samples, 4)
    
    Only uses the first object per image.
    """
    class_labels = []
    bbox_labels = []

    for boxes in y:
        if len(boxes) == 0:
            continue  # Skip images with no objects
        
        first_obj = boxes[0]
        class_id, x_min, y_min, x_max, y_max = first_obj
        class_labels.append(class_id)
        bbox_labels.append([x_min, y_min, x_max, y_max])
    
    return np.array(class_labels), np.array(bbox_labels)
```

```python
def calculate_iou(boxA, boxB):
    # box = [x_min, y_min, x_max, y_max]
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    interArea = max(0, xB - xA) * max(0, yB - yA)
    boxAArea = max(0, boxA[2] - boxA[0]) * max(0, boxA[3] - boxA[1])
    boxBArea = max(0, boxB[2] - boxB[0]) * max(0, boxB[3] - boxB[1])

    unionArea = boxAArea + boxBArea - interArea
    return interArea / unionArea if unionArea != 0 else 0
```

```python
ious = []
for true_box, pred_box in zip(bbox_test, pred_bboxes):
    iou = calculate_iou(true_box, pred_box)
    ious.append(iou)

mean_iou = np.mean(ious)
print(f"Mean IoU on Test Set: {mean_iou:.4f}")
```

```python
def visualize_predictions(model, X, class_labels, bbox_labels, num_samples=5):
    indices = random.sample(range(len(X)), num_samples)

    for idx in indices:
        image = X[idx]
        true_class = class_labels[idx]
        true_bbox = bbox_labels[idx]

        # Predict
        image_input = np.expand_dims(image, axis=0)
        pred_class_probs, pred_bbox = model.predict(image_input, verbose=0)
        pred_class = np.argmax(pred_class_probs[0])
        pred_bbox = pred_bbox[0]

        # Rescale boxes to image size
        img_disp = (image * 255).astype(np.uint8)
        h, w = img_disp.shape[:2]

        true_box = [int(true_bbox[0]*w), int(true_bbox[1]*h), int(true_bbox[2]*w), int(true_bbox[3]*h)]
        pred_box = [int(pred_bbox[0]*w), int(pred_bbox[1]*h), int(pred_bbox[2]*w), int(pred_bbox[3]*h)]

        # Draw boxes
        img_copy = img_disp.copy()
        cv2.rectangle(img_copy, (pred_box[0], pred_box[1]), (pred_box[2], pred_box[3]), (0,255,0), 2)
        cv2.putText(img_copy, f'Pred: {CLASSES[pred_class]}', (pred_box[0], pred_box[1]-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

        cv2.rectangle(img_copy, (true_box[0], true_box[1]), (true_box[2], true_box[3]), (0,0,255), 2)
        cv2.putText(img_copy, f'True: {CLASSES[true_class]}', (true_box[0], true_box[1]+15), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,255), 2)

        # Show image
        plt.figure(figsize=(6,6))
        plt.imshow(cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title('Green = Prediction | Red = Ground Truth')
        plt.show()
```

```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import cv2
import numpy as np

def detect_and_display(image_path):
    # Load and preprocess
    image = cv2.imread(image_path)
    image_resized = cv2.resize(image, (224, 224))
    image_input = image_resized / 255.0
    image_input = np.expand_dims(image_input, axis=0)

    # Predict
    class_probs, bbox_pred = model.predict(image_input, verbose=0)
    class_id = np.argmax(class_probs[0])
    bbox = bbox_pred[0]

    # Denormalize box
    h, w = image.shape[:2]
    x_min = int(bbox[0] * w)
    y_min = int(bbox[1] * h)
    x_max = int(bbox[2] * w)
    y_max = int(bbox[3] * h)

    # Draw prediction
    cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0,255,0), 2)
    cv2.putText(image, CLASSES[class_id], (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0,255,0), 2)

    # Convert to Tkinter-compatible image
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    img_pil = Image.fromarray(image_rgb)
    img_tk = ImageTk.PhotoImage(img_pil)

    # Show in GUI
    label_img.config(image=img_tk)
    label_img.image = img_tk


def browse_image():
    file_path = filedialog.askopenfilename()
    if file_path:
        detect_and_display(file_path)


# GUI setup
root = tk.Tk()
root.title("Face Mask Detection")

btn = tk.Button(root, text="Browse Image", command=browse_image, font=('Arial', 14))
btn.pack(pady=10)

label_img = tk.Label(root)
label_img.pack()

root.mainloop()
```
